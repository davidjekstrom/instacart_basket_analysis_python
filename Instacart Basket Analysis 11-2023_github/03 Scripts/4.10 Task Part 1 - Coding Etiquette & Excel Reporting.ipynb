##### create histogram for income categorized by socioeconomic class

```python
# create bins to represent socioec groups
bins = [0., 28000., 55000., 90000., 150000., max(df_high['income']) * 1.0]
bins_array = np.array(bins)


# create formatted histogram for income
hist_socioec = df_high['income'].plot.hist(bins = bins, title = 'Orders By Socioeconomic Class', xlabel = 'Income', ylabel = 'Orders (in Millions)', edgecolor='black', rot = 45, color = 'lightblue')


# set ticks to match income groupings and relabel
hist_socioec.set_xticks(ticks = [28000, 55000, 90000, 150000], labels = ['$28K', '$55K', '$90K', '$150K'])


# limit x axis to income range
hist_socioec.set_xlim(0, max(df_high['income']))


# add labels to the bins
bin_centers = 0.5 * (bins_array[1:] + bins_array[:-1])
labels = ['Lower', 'Lower middle', 'Middle', 'Upper middle', 'Upper']

for bin_center, label in zip(bin_centers, labels):
    if label == 'Upper':
        plt.text(bin_center, df_high['order_id'][df_high['socioec_class'] == label].count() / 2, label, ha = 'center', rotation = 0)
    elif (label != 'Lower') & (label != 'Lower middle'):
        plt.text(bin_center, df_high['order_id'][df_high['socioec_class'] == label].count() / 2.5, label, ha = 'center', rotation = 90) 
    else:
        plt.text(bin_center, df_high['order_id'][df_high['socioec_class'] == label].count() + 500000, label, ha = 'center', rotation = 90)


# set y-axis to non-scientific notation
plt.gca().ticklabel_format(style='plain', axis='y')


# set y ticks to be divided by 1M
yticks_scaled = hist_socioec.get_yticks() / 1000000
hist_socioec.set_yticks(ticks = hist_socioec.get_yticks(), labels = yticks_scaled)


# show plot
plt.show()
```
