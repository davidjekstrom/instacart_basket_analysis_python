### For Loops

```python
# create list from Set Opener

list_opener = df_gd_shows['Set Opener'].tolist()
```

```python
# print each song in list

for song in list_opener:
    print(song)
```

```python
# print each unique song in list

set_opener = set(list_opener)

for song in set_opener:
    print(song)
```

### Writing Definitions

```python
# define function

def add_num(x,y):
    sum = x * (y + .01)
    if y < 10:
        return add_num(x,sum)
    else:
        return sum
```

```python
# run function

add_num(1,1)
```
