```python
df_prods_revisions_prodid_cleaned.head()
```

There are 2 large outliers. I will replace these with the average price of their respective food type.

```python
# check for other cottage cheese items

df_prods_cotcheese = df_prods_sorted[df_prods_sorted['product_name'].str.contains('Cottage Cheese') == True]
```
