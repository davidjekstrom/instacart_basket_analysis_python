#### 1. Import libraries and data

```python
# Import libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

#### 2. Creating Bar Charts

```python
# plot bar chart for orders_day_of_week

ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
# plot bar chart for orders_day_of_week sorted by day

ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# create bar chart with color

bar = ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

#### 3. Exporting Charts

```python
# set bar chart equal to bar

bar = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# call bar chart

bar.figure
```

```python
# export bar chart

bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

#### 5. Creating Line Charts

```python
# create smaller subset for use in plotting line chart

df = ords_prods_merge[:2000000]
```

```python
# set the seed for random number generator

np.random.seed(4)
```

```python
# create randomized sample using 30/70 split

dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# store 70% of sample dataframe big

big = ords_prods_merge[dev]
```

```python
# store 30% of sample dataframe big

small = ords_prods_merge[~dev]
```

```python
# check big and small dataframes

(len(big) + len(small)) == len(ords_prods_merge)
```

```python
# create subset that only includes needed columns

df_2 = small[['orders_day_of_week','prices']]
```

```python
# creat line chart from df_2

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# export line chart

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_sample_orders_dow.png'))
```
