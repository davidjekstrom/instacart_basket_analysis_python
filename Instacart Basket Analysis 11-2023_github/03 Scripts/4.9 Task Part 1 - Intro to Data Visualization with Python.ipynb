#### Part 1 Question 5

Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

The stastics don't immediately signal inconsistencies.

```python
# check max price row

df_cust[df_cust['income'] == df_cust['income'].max()]
```

```python
# check for mixed-type columns in a df_cust

for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].map(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# change first_name data type to string

df_cust['first_name'] = df_cust['first_name'].astype('str')
```

```python
# re-check for mixed-type columns in a df_cust

for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].map(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# check for missing values and return count for each column

df_cust.isnull().sum()
```

#### Part 1 Question 6

Combine your customer data with the rest of your prepared Instacart data. (Hint: Make sure the key columns are the same data type!)

```python
# import ords_prods_merged_derived_final.pkl

df_ords_prods = pd.read_pickle(os.path.join(path,'02 Data','Prepared Data','ords_prods_merged_derived_final.pkl'))
```

```python
# import df_cust.pkl

df_cust = pd.read_pickle(os.path.join(path,'02 Data','Prepared Data','customers_checked_.pkl'))
```

```python
# check df_ords_prods head

df_ords_prods.head()
```

```python
# remove jupyter's columns limit

pd.options.display.max_columns = None
```

```python
# check value counts of _merge

df_ords_prods['_merge'].value_counts()
```

```python
# drop _merge col that remained from previous merge

df_ords_prods = df_ords_prods.drop('_merge', axis=1)
```

```python
# check shape

df_ords_prods.shape
```

```python
# check df_cust head

df_cust.head()
```
