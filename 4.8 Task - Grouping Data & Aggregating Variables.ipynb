#### Question 4

Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# create a max order column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check output

ords_prods_merge.head()
```

```python
# create loyalty flag for loyal customers

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# create loyalty flag for regular customers

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# create loyalty flag for new customers

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check value counts of loyalty_flag

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# check output

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head()
```

#### Question 5

The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# check prices statistics for by loyalty category

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# import scipy

from scipy import stats
```

```python
# define loyal group prices as variable for significance test
loyal_prices = ords_prods_merge[ords_prods_merge['loyalty_flag'] == 'Loyal customer']['prices']
```

```python
# define non loyal group prices as variable for significance test

nonloyal_prices = ords_prods_merge[ords_prods_merge['loyalty_flag'] != 'Loyal customer']['prices']
```

```python
# perform t-test 

t_statistic, p_value = stats.ttest_ind(loyal_prices, nonloyal_prices, equal_var=False)
```

```python
# check p-value

p_value < .05
```

##### Question 5 Answers

The statistics of prices for each of the 3 loyalty groups are remarkably similar with one exception. Loyal customers purchase products at a slightly lower overall average price than both new and regular customers. The difference is small but does reflect a statistically significant behavior difference among loyal customers.

The difference between the average price of new and regular customers is slightly higher for new customers but by a nominal amount.

#### Question 6

The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

- If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
- If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# create an average_price column

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# create spending flag for low spenders

ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# create spending flag for high spenders

ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# check value counts of spending_flag

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# check output

ords_prods_merge[['user_id', 'spending_flag', 'average_price']].head(100)
```

#### Question 7

In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
- If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
- If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
- If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# create an days_since_prior_median column

ords_prods_merge['days_since_prior_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# create order_freq_flag for non-frequent customer

ords_prods_merge.loc[ords_prods_merge['days_since_prior_median'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
# create order_freq_flag for Regular customer

ords_prods_merge.loc[(ords_prods_merge['days_since_prior_median'] > 10) & (ords_prods_merge['days_since_prior_median'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
# create order_freq_flag for Frequent customer

ords_prods_merge.loc[ords_prods_merge['days_since_prior_median'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
# check value counts of order_freq_flag

ords_prods_merge['order_freq_flag'].value_counts(dropna = False)
```

```python
# check output

ords_prods_merge[['user_id', 'order_freq_flag', 'days_since_prior_median']].head(20)
```
